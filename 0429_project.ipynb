```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='NanumBarunGothic')    # 코랩에서 사용
#plt.rc('font', family='Malgun Gothic')    # 윈도우에서 사용
plt.rcParams['axes.unicode_minus'] = False
```

```python
filename= '과실생산량_성과수미과수__20240429192116.csv'
df= pd.read_csv(filename, encoding='utf-8')

df.index= df.시도별
df.drop('시도별',axis=1, inplace=True)
df.head()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)

x = df.columns
y = list(df.iloc[1].apply(int))

print(x)
print(y)
```

```python
# 2.막대 그래프 그리기
plt.bar(x,y)
plt.title('사과 생산량')     # 그래프 제목
plt.xlabel('년도')     # x축 이름
plt.ylabel('생산량')# y축 이름
plt.xticks(rotation=45)     # X축 레이블 회전
plt.savefig('년도별사과생산량.png')
plt.show()
```

```python
# 선 그래프
plt.rcParams['figure.figsize']=(13,10)

for i in range(10,18):
    df.iloc[i]= df.iloc[i].str.replace('-','0')
    y = list(df.iloc[i].apply(int))
    if df.index[i] == '경상북도' :
        kb_y=df.iloc[i].apply(int)
    plt.plot(x,y,'o',linestyle='solid',label=df.index[i])
plt.grid()
plt.legend()
plt.xticks(rotation=45)
plt.savefig('년도별지역별사과생산량.png')
plt.show()
```

```python
df80= list(df['1980'][10:18].apply(int))
label=list(df.index[10:18])
print(df80)
print(label)
fig, ax = plt.subplots()
# 원그래프 그리기
ax.pie(df80, labels=label, autopct='%1.1f%%',textprops={'fontsize': 18})
ax.axis('equal')  # 원형 비율 유지
ax.legend(prop={'size': 20},loc='center left')
plt.savefig('1980년각지사과생산량.png')
plt.show()
```

```python
df23= list(df['2023'][10:18].apply(int))
label=list(df.index[10:18])
print(df23)
print(label)
fig, ax = plt.subplots()
# 원그래프 그리기
ax.pie(df23, labels=label, autopct='%1.1f%%',textprops={'fontsize': 18})
ax.axis('equal')  # 원형 비율 유지
plt.savefig('2023년각지사과생산량.png')
plt.show()
```

```python

filename= 'OBS_ASOS_MNH_20240423121211.csv'
df2= pd.read_csv(filename, encoding='cp949')
df2.columns = ['지점', '지점명', '일시', '평균기온', '최고기온', '최저기온', '평균현지기압','평균상대습도', '월합강수량', '평균풍속','일조시간', '최심적설']
df2.insert(3,'년도',df2['일시'].str[:4])

df2 = df2[df2['년도'] != '2024']
df2 = df2[df2['년도'] >= '1980']

kyungbuk_df=df2.query("지점명 in ('안동','구미','김천','문경','상주','영주','영천','예천','의성','청송')")
kb_mean=kyungbuk_df.groupby(['지점명','년도']).mean(numeric_only=True)
```

```python
# 경남 지점 연도별 그룹핑하기
df_kn = df2.query(" 지점명 in ('안동','구미','김천','문경','상주','영주','영천','예천','의성','청송') ")
df_kn = df_kn.groupby(['년도']).aggregate({'평균기온':'mean'}).reset_index()
df_kn

X = df_kn.년도
Y = df_kn.평균기온

plt.rcParams['figure.figsize']=(13,10)
# 경남 지점 평균 구하기
plt.plot(X, Y, '^', linestyle='solid', label='평균기온(°C)')

plt.title('경북 지점 평균기온', fontsize=20)
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.xticks(rotation=45)
plt.grid()
plt.savefig('경북년도별평균기온.png')
plt.show()
```
