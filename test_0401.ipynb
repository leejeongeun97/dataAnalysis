```python
coin= int(input("동전으로 교환할 금액을 입력하세요"))
print('-'*30)
coin500= coin // 500
coin -= (coin500*500)
coin100= coin // 100
coin -= (coin100*100)
coin50= coin // 50
coin -= (coin50*50)
coin10= coin // 10
coin -= (coin10*10)
ex=coin%10

print("500원짜리 ==> %d개" % coin500)
print("100원짜리 ==> %d개" % coin100)
print("50원짜리 ==> %d개" % coin50)
print("10원짜리 ==> %d개" % coin10)
print("잔돈 ==> %d원" % ex)



```
